```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from ydata_profiling import ProfileReport
```

```python
df=pd.read_csv('dataset.csv')
```

```python
df.describe(include='all')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
for col in df.columns:
    print(col)
    print(df[col].unique())
#getting the unique value
```

```python
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
for col in numeric_cols:
    plt.figure(figsize=(10,4))
    sns.histplot(df[col], kde=True)
    plt.title(col)
    plt.show()
#knowing the distributions
```

```python
for col in numeric_cols:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=df[col])
    plt.title(col)
    plt.show()
#boxplotting to find outliers
```

```python
corr = df[numeric_cols].corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('correlation')
plt.show()
#correlation matrix
```

```python
sns.pairplot(df[numeric_cols])
plt.show()
#plotting columns against eachother
```

```python
print(df.groupby('Sex')['Survived'].mean())
print(df.groupby('Pclass')['Survived'].mean())
#females and passenger class have higher survival rate
```

```python
sns.histplot(data=df, x='Age', hue='Survived', multiple='stack')
#ages against survived
```

```python
sns.boxplot(x='Survived', y='Fare', data=df)
#high fare implies higher survival rate
```

```python
df.groupby('Embarked')['Survived'].mean()
#embarked at C have higher survival rate
```

```python
df['FamilySize'] = df['SibSp'] + df['Parch']
sns.barplot(x='FamilySize', y='Survived', data=df)
#optimal family size 3 has high survival rate and Alone persons are less survived
```

```python
df['HasCabin'] = df['Cabin'].notnull().astype(int)
df.groupby('HasCabin')['Survived'].mean()
#missing cabin values implies higher survival rate
```

```python
sns.barplot(x='Pclass', y='Survived', hue='Sex', data=df)
#higher class implies higher survival rate
```

```python
df['AgeBin'] = pd.cut(df['Age'], bins=[0, 10, 18, 40, 60, 80])
df.groupby('AgeBin')['Survived'].mean()
#younger peoples are more survived
```

```python
profile = ProfileReport(df, title="Data Profile Report", explorative=True)
profile.to_file("report.html")
```
